## Be able to separate by Treatment Type

```python
summary = summary[(summary.total_cost > 0) & (summary.duration < 100)]
expLabels = summary.groupby(['n_part_requested', 'experiment_type'])['session_code'].unique()
expLabels.xs(10,level=0).iloc[0]
label_dict = {s[1]['session_code'] : (s[1]['n_part_requested'], s[1]['experiment_type']) for s in summary.iterrows()}
```

## Measurable Inputs; need to show all available from study as well.

```python
instructions_anagrams = instructions_anagrams[instructions_anagrams['session.code'].isin(label_dict.keys())]
exp_type = instructions_anagrams.apply(lambda x: label_dict[x['session.code']][1], axis = 1)
n_part = instructions_anagrams.apply(lambda x: label_dict[x['session.code']][0], axis = 1)
instructions_anagrams['n_part_requested'] = n_part
instructions_anagrams['experiment_type'] = exp_type
```
